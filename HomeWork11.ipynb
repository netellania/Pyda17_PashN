```python
# Задание 1
#Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
#- оценка 2 и меньше - низкий рейтинг
#- оценка 4 и меньше - средний рейтинг
#- оценка 4.5 и 5 - высокий рейтинг
#Результат классификации запишите в столбец class
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
films = ratings.merge(movies, on='movieId', how='left')

def Class(row):
    if ( row['rating']== 2.0 or  row['rating'] < 2.0):
        return 'низкий рейтинг'
    if ( row['rating']== 4.0 or (row['rating']>2.0 and row['rating']<4.0)):
        return 'средний рейтинг'
    if ( row['rating']== 5.0 or row['rating']==4.5):
        return 'высокий рейтинг'
films['Class'] = films.apply(Class, axis=1)    
films.head()
```

```python
films.Class.value_counts()
```

```python
#Используем файл keywords.csv.
#Необходимо написать гео-классификатор,
#который каждой строке сможет выставить географическую принадлежность
#определенному региону. 
geo = pd.read_csv('keywords.csv')
geo.head()
```

```python
geo = pd.read_csv('keywords.csv')
def region(column):
    if 'москва' in column['keyword'] or 'тула'in column['keyword'] or  'ярославль'in column['keyword']:
        return 'Центр'
    elif 'петербург' in column['keyword'] or 'псков'in column['keyword'] or  'мурманск'in column['keyword']:
        return 'Северо-Запад'
    elif 'владивосток' in column['keyword'] or 'сахалин'in column['keyword'] or  'хабаровск'in column['keyword']:
        return 'Дальний Восток'
    return 'undefined'
geo['region'] = geo.apply(region, axis=1)    
geo.head()
```

```python
geo.region.value_counts()
```
