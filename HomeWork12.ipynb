```python
import pandas as pd
```

```python
# Задача 1
#Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:
#если источник traffic_source равен yandex или google, то в source_type ставится organic
#для источников paid и email из России - ставим ad
#для источников paid и email не из России - ставим other
#все остальные варианты берем из traffic_source без изменений
log = pd.read_csv('visit_log.csv', sep=';')
log_new=log.assign(source_type=log.traffic_source)
log_new.loc[log_new.traffic_source.isin(['yandex', 'google']), 'source_type' ] = 'organic'
log_new.loc[(log_new.region == 'Russia') & (log_new.traffic_source.isin(['paid', 'email'])), 'source_type' ] = 'ad'
log_new.loc[(log_new.region != 'Russia') & (log_new.traffic_source.isin(['paid', 'email'])), 'source_type' ] = 'other'
log_new.head(20)
```

```python
# Задача 2
#В файле URLs.txt содержатся url страниц новостного сайта. 
#Вам необходимо отфильтровать его по адресам страниц с текстами новостей. 
#Известно, что шаблон страницы новостей 
#имеет внутри url следующую конструкцию: /, затем 8 цифр, затем дефис. 
#Выполните следующие действия:
#Прочитайте содержимое файла с датафрейм
#Отфильтруйте страницы с текстом новостей, используя метод str.contains и регулярное выражение
#в соответствии с заданным шаблоном
news = pd.read_csv('URLs.txt', sep=';')
news[news.url.str.contains('/\d{8}-')].head()
```

```python
# Задача 3
#Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
#Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
#Под временем жизни понимается 
#разница между максимальным и минимальным значением столбца timestamp 
#для данного значения userId.
ratings = pd.read_csv('ml-latest-small/ratings.csv')
df_ratings=ratings[ratings.movieId>100]
def groupby_function(df_ratings):
    return df_ratings.timestamp.max() - df_ratings.timestamp.min()
df_ratings.groupby('userId').apply(groupby_function)
```

```python
# Задача 3
# Дана статистика услуг перевозок клиентов компании по типам (см. файл с кодом занятия). 
#Необходимо сформировать две таблицы:
#1.Таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
#2.Аналогичную таблицу по типам выручки с указанием адреса клиента
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd = rzd[['client_id', 'rzd_revenue']]
rzd
```

```python

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

auto = auto[['client_id', 'auto_revenue']]
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air = air[['client_id', 'air_revenue']]
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)

client_base = client_base[['client_id', 'address']]
client_base
```

```python
#Промежуточное соединение
joi1 = rzd.merge(auto, on='client_id', how='outer')
joi1.head(10)
```

```python
joi2 = joi1.merge(air, on='client_id', how='outer')
joi2.head(10)
```

```python
joi3 = rzd.merge(auto, on='client_id', how='outer').merge(air, on='client_id', how='outer')
joi3.head(10)
```

```python
joi4 = client_base.merge(joi3, on='client_id', how='outer')
joi4.head(10)
```
